```python
# Selecting Duration (ms) cloumn to detect outliers in our data
numerical_columns = ['Duration (ms)']  

# Create a box plot for each numerical column
plt.figure(figsize=(10, 6))
data[numerical_columns].boxplot()
plt.title('Box Plot of Numerical Columns')
plt.ylabel('Values')
plt.show()
```

```python
## Top 10 songs of all time
# Sort the data by popularity in descending order
top_songs = data.sort_values(by='Popularity', ascending=False).head(10)

# Plot a bar chart for the top 10 songs
plt.figure(figsize=(12, 6))
plt.barh(top_songs['Track Name'], top_songs['Popularity'], color='skyblue')
plt.xlabel('Popularity')
plt.ylabel('Song')
plt.title('Top 10 Songs of All Time')
plt.gca().invert_yaxis()  
plt.show()
```

```python
# Group the data by artists and sum up the popularity scores
top_artists = data.groupby('Artist')['Popularity'].sum().reset_index()

# Sort the data by total popularity in descending order
top_artists = top_artists.sort_values(by='Popularity', ascending=False).head(10)

# Plot a bar chart for the top 10 artists
plt.figure(figsize=(12, 6))
plt.barh(top_artists['Artist'], top_artists['Popularity'], color='green')
plt.xlabel('Total Popularity')
plt.ylabel('Artist')
plt.title('Top 10 Artists by Total Popularity')
plt.gca().invert_yaxis()  
plt.show()
```

```python
# Sort the dataset by duration in descending order
data_sorted_by_duration = data.sort_values(by='Duration (ms)', ascending=False)

# Select the top 10 songs with the longest duration
top_10_duration_songs = data_sorted_by_duration.head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_duration_songs['Track Name'], top_10_duration_songs['Duration (ms)'], color='skyblue')
plt.xlabel('Track Name')
plt.ylabel('Duration (ms)')
plt.title('Top 10 Songs with Longest Duration')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Duration (ms)'], data['Popularity'], color='blue', alpha=0.5)
plt.title('Relationship between Popularity and Duration')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python

# Calculate the mean popularity for each artist
artist_popularity = data.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
artist_popularity.plot(kind='bar', color='purple')
plt.title('Top 10 Artists by Mean Popularity of Songs')
plt.xlabel('Artist')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
